```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Financial Analysis Automation Started")
```

```python
financial_data = {
    "Year": [2021, 2022, 2023, 2024],
    
    "Revenue": [1000, 1200, 1500, 1800],
    
    "Cost_of_Goods_Sold": [600, 700, 850, 950],
    
    "Net_Profit": [120, 180, 250, 320],
    
    "Assets": [3000, 3500, 4000, 4600],
    
    "Liabilities": [1500, 1600, 1700, 1800]
}


df = pd.DataFrame(financial_data)

df
```

```python
# Calculate financial indicators

df["Gross_Profit"] = (
    df["Revenue"] - df["Cost_of_Goods_Sold"]
)

df["Gross_Margin"] = (
    df["Gross_Profit"] / df["Revenue"]
)

df["Profit_Margin"] = (
    df["Net_Profit"] / df["Revenue"]
)

df["Debt_Ratio"] = (
    df["Liabilities"] / df["Assets"]
)

df
```

```python
# Calculate yearly growth

df["Revenue_Growth"] = (
    df["Revenue"].pct_change()
)

df["Profit_Growth"] = (
    df["Net_Profit"].pct_change()
)

df
```

```python
# Convert ratios into percentage format

df_style = df.copy()

percentage_columns = [
    "Gross_Margin",
    "Profit_Margin",
    "Debt_Ratio",
    "Revenue_Growth",
    "Profit_Growth"
]


for col in percentage_columns:
    df_style[col] = (
        df_style[col] * 100
    ).round(2).astype(str) + "%"


df_style
```

```python
# Automated financial interpretation

latest_year = df.iloc[-1]

report = f"""
Financial Analysis Report

Period:
{df['Year'].min()} - {df['Year'].max()}


Key Findings:

1. Revenue:
Revenue increased from {df['Revenue'].iloc[0]} 
to {latest_year['Revenue']}.

2. Profitability:
Net profit increased from {df['Net_Profit'].iloc[0]}
to {latest_year['Net_Profit']}.

3. Margin:
Gross margin improved to 
{latest_year['Gross_Margin']:.2%}.

4. Financial Risk:
Debt ratio decreased to
{latest_year['Debt_Ratio']:.2%}.

Overall:
The company shows improving profitability
and decreasing financial leverage.
"""

print(report)
```

```python
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet


pdf_path = "../reports/financial_analysis_report.pdf"


doc = SimpleDocTemplate(pdf_path)


styles = getSampleStyleSheet()

content = []


report_text = f"""
Financial Analysis Report


Period:
{df['Year'].min()} - {df['Year'].max()}


Financial Performance:

Revenue increased from
{df['Revenue'].iloc[0]}
to
{df['Revenue'].iloc[-1]}


Net Profit increased from
{df['Net_Profit'].iloc[0]}
to
{df['Net_Profit'].iloc[-1]}



Profitability:

Gross Margin:
{df['Gross_Margin'].iloc[-1]:.2%}


Profit Margin:
{df['Profit_Margin'].iloc[-1]:.2%}



Financial Risk:

Debt Ratio:
{df['Debt_Ratio'].iloc[-1]:.2%}



Recommendation:

Continue improving operational efficiency
and maintain healthy financial structure.
"""


for line in report_text.split("\n"):

    content.append(
        Paragraph(line, styles["Normal"])
    )

    content.append(
        Spacer(1,12)
    )


doc.build(content)


print("PDF report generated successfully")
```
